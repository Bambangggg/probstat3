```python
import pandas as pd

data_bambang = pd.read_csv("/Users/bambangaji/Downloads/data_tif.csv")

# Menampilkan data
print(data_bambang)
```

```python
subset_data = data_bambang[['Nama', 'Tinggi Badan']]

# Menampilkan subset data
print(subset_data)
```

```python
data_bambang.loc[data_bambang['Tinggi Badan'] > 160, 'Tinggi Badan'] = 'Tinggi'
data_bambang.loc[data_bambang['Tinggi Badan'] != 'Tinggi', 'Tinggi Badan'] = 'Pendek'

# Menampilkan data setelah perubahan
print(data_bambang)
```

```python
data_bambang = data_bambang.assign(Jurusan="Infor20")

# Menambahkan variabel kolom baru "Fakultas" dengan isi "FTI"
data_bambang = data_bambang.assign(Fakultas="FTI")

# Menampilkan data setelah penambahan variabel kolom baru
print(data_bambang)
```

```python
data_bambang = data_bambang.drop(columns=['Fakultas'])

print(data_bambang)
```

```python
data_bambang['Nama_Gender'] = data_bambang['Nama'] + '_' + data_bambang['Gender']

# Gabung kolom Angkatan dan Tinggi Badan
data_bambang['Angkatan_TinggiBadan'] = data_bambang['Angkatan'].astype(str) + '_' + data_bambang['Tinggi Badan'].astype(str)

# Gabungkan 2 variabel kolom gabungan di atas
data_bambang['Gabungan'] = data_bambang['Nama_Gender'] + '_' + data_bambang['Angkatan_TinggiBadan']

# Menampilkan data setelah penggabungan
print(data_bambang)
```

```python
gabungan_1 = data_bambang.iloc[0:5].sum()

# Gabung baris 25-30
gabungan_2 = data_bambang.iloc[24:30].sum()

# Gabungkan 2 variabel di atas
gabungan_akhir = gabungan_1 + gabungan_2

# Menampilkan hasil gabungan akhir
print(gabungan_akhir)
```

```python
data_bambang_sorted = data_bambang.sort_values(by='Waktu Perjalanan')

# Menampilkan data setelah diurutkan
print(data_bambang_sorted)
```
